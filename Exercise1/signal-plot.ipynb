```python
import numpy as np
import matplotlib.pyplot as plt
print("NumPy version:", np.__version__)
```

## Creating Fake Sensor Data

```python
N_SAMPLES = 500
```

```python
input_range = np.linspace(0,2*np.pi, N_SAMPLES, dtype=np.float32) # x values
```

```python
signal = np.sin(input_range) # y values where y = sin(x)
```

```python
noise = np.random.randn(N_SAMPLES) # N_SAMPLES random values from the normal distribution
```

```python
assert noise.shape == input_range.shape # check if noise and input have same shape/size
```

```python
noisy_signal = signal + noise/5 # add noise
```

```python
plt.plot(input_range, noisy_signal, 'b.', alpha=0.5)
plt.plot(input_range, signal, 'r-', linewidth=4)
plt.legend(['Sensor readings', 'Truth'])
plt.xlabel('Time')
plt.ylabel('Value of thing we care about')
plt.show()
```

## Filtering Noise

```python
del signal
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
filtered = lowess(noisy_signal, input_range, frac=0.1)
```

```python
plt.plot(input_range, noisy_signal, 'b.', alpha=0.5)
plt.plot(filtered[:,0], filtered[:,1], 'r-', linewidth=4)
plt.legend(['Sensor readings', 'Reconstructed signal'])
plt.show()
```
