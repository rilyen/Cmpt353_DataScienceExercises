```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# data = pd.read_csv("data-1.csv")
```

```python
for i in range(1,7):
    filename = "data-" + str(i) + ".csv"
    data = pd.read_csv(filename)
    d_mean = data.mean()
    d_std = data.std()
    d_min = data.min()
    d_max = data.max()
    d_r = stats.linregress(data['x'],data['y']).rvalue
    print("SUMMARY STATISTICS: " + filename)
    print("mean of x: ", end="")
    print(d_mean['x'])
    print("mean of y: ", end="")
    print(d_mean['y'])
    print("standard deviation of x: ", end="")
    print(d_std['x'])
    print("standard deviation of y: ", end="")
    print(d_std['y'])
    print("range of x: ", end="")
    print(d_min['x'], end="")
    print(" to ", end="")
    print(d_max['x'])
    print("range of y: ", end="")
    print(d_min['y'], end="")
    print(" to ", end="")
    print(d_max['y'])
    print("correlation coefficient between x and y: ", end="")
    print(d_r)
    print("\n")
    figurename = "data-" + str(i) + ".png"
    plt.figure(figsize=(10, 6))
    plt.scatter(data['x'], data['y'], marker='o', color='blue', alpha=0.7)
    plt.title(f"Scatter plot of {filename}")
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.savefig(figurename)
```
